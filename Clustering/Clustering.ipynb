```python
X = data.iloc[:,:].values
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
```

```python
hc = AgglomerativeClustering(n_clusters=5,linkage='ward')
```

```python
predict_hc =            hc.fit_predict(X)
```

```python
print(predict_hc)
```

```python
plt.scatter(X[predict_hc==0,0],X[predict_hc==0,1],c='r',label='cluster1')
plt.scatter(X[predict_hc==1,0],X[predict_hc==1,1],c='g',label='cluster2')
plt.scatter(X[predict_hc==2,0],X[predict_hc==2,1],c='b',label='cluster3')
plt.scatter(X[predict_hc==3,0],X[predict_hc==3,1],c='y',label='cluster4')
plt.scatter(X[predict_hc==4,0],X[predict_hc==4,1],c='pink',label='cluster5')
```
