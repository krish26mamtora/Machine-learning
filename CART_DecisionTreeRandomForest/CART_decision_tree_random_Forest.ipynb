```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
class_name=iris.target_names
```

```python
class_name
```

```python
features = iris.feature_names
```

```python
features
```

```python
data = pd.DataFrame({'sepal length':iris.data[:,0],
                     'sepal width':iris.data[:,1],
                     'petal length':iris.data[:,2],
                     'petal width':iris.data[:,3],
                     'species':iris.target})
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data[['sepal length','sepal width','petal length','petal width']]
y=data[['species']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(n_estimators=10)
forest.fit(x_train,y_train)
```

```python
y_pred=forest.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```

```python
import pydot
from sklearn.tree import export_graphviz
```

```python
tree=forest.estimators_[4]
```

```python
export_graphviz(tree,out_file="/content/tree5.dot",feature_names=features)
```

```python
(graph,)=pydot.graph_from_dot_file("/content/tree5.dot")
```

```python
graph.write_png("/content/tree5.png")
```
